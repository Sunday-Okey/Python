## Big O

Big O is a way of comparing two sets of code

```python
mylist = [2, 3, 7, 10, 4]
```

```python
def sum_to_target(lst, target):
    myset = set()
    
    for num in lst:
        complement = target - num
        if complement in myset:
            return num, complement
        myset.add(num)
```

```python
sum_to_target(mylist, 13)
```

Time complexity is measured in the number of operations taken to complete something
