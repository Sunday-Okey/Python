### Modifying Lists

```python
myList = [1,2,3,4]
myList.append(5)
print(myList)
```

```python
myList.insert(3, 'a new value')
print(myList)
```

```python
myList.remove('a new value')
```

```python
myList
```

```python
myList.pop()
```

```python
while len(myList):
    print(myList.pop())
```
