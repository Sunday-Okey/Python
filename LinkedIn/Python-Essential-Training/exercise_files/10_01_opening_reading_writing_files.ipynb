```python
with open("10_02_us.csv", 'r') as f:
    reader = csv.DictReader(f, delimiter='\t')
    data = list(reader)
```

```python
print(data[0])
```

```python
zipcode = []
for line in data:
    zipcode.append(int(line['postal code']))

```

```python
primes = []
# zipcode is a list of zip codes
for n in zipcode:
    mylist = [2]
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            break
    else:
        primes.append(n)
```

```python
primes = []

for n in zipcode:
    n = int(n)          # in case zip codes are strings
    if n < 2:
        continue

    is_prime = True
    for p in primes:
        if p * p > n:   # same as p > sqrt(n) but faster
            break
        if n % p == 0:
            is_prime = False
            break

    if is_prime:
        primes.append(n)
```
