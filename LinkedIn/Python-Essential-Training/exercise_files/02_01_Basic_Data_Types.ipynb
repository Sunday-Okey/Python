```python
k = int(input("How many Fibonacci Sequence numbers do you want?"))
n = 1
x, y = 0, 1
print(0, 1, sep='\n')

while n <= k:
    p = x + y
    print(p)
    x, y = y, p
    n += 1
```

```python
k = int(input("Enter a number: "))
n = 1
total = 1
while n <= k:
    total *= n
    n += 1
print(total)
```

```python
def factorial(k):
    n = 1
    total = 1
    while n <= k:
        total *= n
        n += 1
    return total
```

```python
factorial(6)
```
